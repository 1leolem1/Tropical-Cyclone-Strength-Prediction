```python
# prompt: can you create a dataframe where df['TD9636_STAGE'] is not null

import pandas as pd

# Load dataset, treating various representations of missing values
df = pd.read_csv("ibtracs.csv", na_values=["NAN", "nan", "NaN", " ", None])

# Create a new DataFrame where 'TD9636_STAGE' is not null
df_filtered = df[df['TD9636_STAGE'].notnull()]

# Display or use the filtered DataFrame
df_filtered['TD9636_STAGE']
```

```python

Final_ibtracs['USA_ATCF_ID'] = Final_ibtracs['USA_ATCF_ID'].fillna('')
```

```python


Final_ibtracs['STORM_SPEED'] = pd.to_numeric(Final_ibtracs['STORM_SPEED'], errors='coerce')

mean_wind = Final_ibtracs['STORM_SPEED'].mean()

Final_ibtracs['STORM_SPEED'].fillna(mean_wind, inplace=True)



Final_ibtracs['STORM_DIR'] = pd.to_numeric(Final_ibtracs['STORM_DIR'], errors='coerce')

mean_wind = Final_ibtracs['STORM_DIR'].mean()

Final_ibtracs['STORM_DIR'].fillna(mean_wind, inplace=True)
```
